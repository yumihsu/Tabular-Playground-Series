## data engine

```python
def feature_engine(df,type = 'none'):
    # datetime_processing 
    datatime = df['date_time']
    df['hour'] = pd.to_datetime(df['date_time']).dt.hour
    df['month'] = pd.to_datetime(df['date_time']).dt.month
    df = df.drop(['date_time'],axis = 1)
    # mean of senser
    df['125_sum'] = (df['sensor_1'] + df['sensor_2'] + df['sensor_5'])/3
    # remove 6675 - 6775
    if type == 'train' :
        df = df.drop(df.index[[i for i in range(6675,6775)]],axis = 0).reset_index(drop = True)
    # adjust mean_deg_c
    gap = 1

    mean_deg_c = df['deg_C'].copy()
    change = 0
    for i in range(1,len(mean_deg_c)-1):
        #increase too much
        if mean_deg_c[i-1] - mean_deg_c[i] >  gap:
            mean_deg_c[i] = mean_deg_c[i-1] - gap
            change +=1
        #decrease too much    
        if mean_deg_c[i-1] - mean_deg_c[i] <  gap*-1:
            mean_deg_c[i] = mean_deg_c[i-1] + gap
            change +=1
    df['mean_deg_c'] = mean_deg_c
    df = df.drop(['deg_C','relative_humidity','absolute_humidity'],axis = 1)
    return df ,datatime
```
